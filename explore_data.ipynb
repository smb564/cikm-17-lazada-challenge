```python
import pandas as pd
```

```python
x = pd.read_csv("training/data_train.csv", header=None)
clarity = pd.read_csv("training/clarity_train.labels")
con = pd.read_csv("training/conciseness_train.labels")
```

```python
clarity.columns = ["clarity"]
con.columns = ["conciseness"]
data = x.join(clarity).join(con)
```

```python
data.loc[data.clarity==0].count()
```

```python
print("Clarity from total: ", int(data['clarity'].sum())," out of ", len(data))
print("Conciseness from total: ", int(data['conciseness'].sum())," out of ", len(data))
```

```python
print("Concise given clarity: ", len(data[data.clarity == 1][data.conciseness == 1]), " out of ", len(data[data.clarity == 1]))
```

## Key observation
According to the training data, if a title is concise it must have clarity
